Task 11: Time-based Visualization

Simulate time-based churn by grouping tenure buckets and plotting churn rate

```python
df_plot = pd.read_csv("Telco-Customer-Churn.csv")
df_plot["TotalCharges"] = pd.to_numeric(df_plot["TotalCharges"], errors="coerce")
df_plot.dropna(inplace=True)
df_plot.replace({"No internet service": "No", "No phone service": "No"}, inplace=True)
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ["0-12", "12-24", "24-36", "36-48", "48-60", "60-72"]
df_plot["tenure_group"] = pd.cut(df_plot["tenure"], bins=bins, labels=labels, include_lowest=True)
if df_plot["Churn"].dtype == "object":
    df_plot["Churn_numeric"] = df_plot["Churn"].map({"No": 0, "Yes": 1})
else:
    df_plot["Churn_numeric"] = df_plot["Churn"]
churn_rate_by_bucket = df_plot.groupby("tenure_group")["Churn_numeric"].mean() * 100
print(churn_rate_by_bucket)
plt.figure(figsize=(8,5))
churn_rate_by_bucket.plot(kind="bar", color="orange")
plt.title("Churn Rate by Tenure Group")
plt.xlabel("Tenure Group (months)")
plt.ylabel("Churn Rate (%)")
plt.ylim(0, 100)
plt.show()
```

# Part A: Data Loading and Basic Inspection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler , LabelEncoder
import seaborn as sns
```

# Part B: Data Cleaning and Preprocessing

Task 3: Handle Missing and Incorrect Data

Check for missing values using isnull().sum().

```python
df.isnull().sum()
```

Impute or drop missing values based on proportion and domain knowledge.

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df["sale_status"] = df["soldOn"].apply(lambda x: "Sold" if pd.notnull(x) else "For Sale")
print("\nSale status counts:")
print(df["sale_status"].value_counts())
```

```python
df.isnull().sum()
```

Task 4: Drop Unnecessary Columns

Drop irrelevant identifiers such as PIN, Property Index Number, Transaction ID
if present.

```python
if "text" in df.columns:
    df.drop(columns="text", inplace=True)
id_cols = ["PIN", "Property Index Number", "Transaction ID", "transaction_id", "property_id"]
cols_to_drop = [col for col in id_cols if col in df.columns]
df.drop(columns=cols_to_drop, inplace=True)
```

Task 5: Standardize Categorical Variables

Clean and unify entries for columns like Property Type, Building Class,
Community Area.

```python
df['type'] = df['type'].astype(str).str.strip().str.title()
```

Task 6: Encode Categorical Variables

Label Encode binary columns (e.g., Land Type: Residential/Commercial).

```python
df['sale_status'] = df['sale_status'].map({'For Sale': 0, 'Sold': 1})
```

One-Hot Encode Property Type, Building Class.

```python
df = pd.get_dummies(df, columns=['type'], drop_first=True)
```

Task 7: Feature Scaling

Normalize or scale continuous variables such as Sale Price, Building Area, Lot
Area, Year Built.

```python
scale_cols = ['lastSoldPrice', 'sqft', 'lot_sqft']
scale_cols = [col for col in scale_cols if col in df.columns]
df[scale_cols] = StandardScaler().fit_transform(df[scale_cols])
```

Task 11: Time-based Visualization

Lineplot: Average Sale Price trend over time (monthly or quarterly)

```python
df['soldOn'] = pd.to_datetime(df['soldOn'], errors='coerce')
df['Month'] = df['soldOn'].dt.to_period('M')
monthly_avg = df.groupby('Month')['lastSoldPrice'].mean()
monthly_count = df.groupby('Month')['lastSoldPrice'].count()
plt.plot(monthly_avg.index.astype(str), monthly_avg.values, marker='o', color='orange')
for x, y in zip(monthly_avg.index.astype(str), monthly_avg.values):
    plt.text(x, y, f"{y:,.0f}", ha='center', va='bottom', fontsize=8)
plt.title("Average Sale Price Over Time"); plt.xticks(rotation=45); plt.ylabel("Average Price")
plt.show()
```

Barplot: Number of transactions per month in 2024

```python
plt.bar(monthly_count.index.astype(str), monthly_count.values, color='skyblue', edgecolor='black')
for x, y in zip(monthly_count.index.astype(str), monthly_count.values):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=8)
plt.title("Number of Transactions Per Month (2024)"); plt.xticks(rotation=45); plt.ylabel("Transactions")
plt.show()
```

# Part D: Feature Engineering

Task 12: Create New Features

Property Age: 2024 - Year Built

Price per SqFt: Sale Price / Building Area

IsLargeLot: Indicator if Lot Area > threshold (e.g., 10,000 sq ft)

```python
df['Property Age'] = 2024 - df['year_built']
df['Price per SqFt'] = df['lastSoldPrice'] / df['sqft']
df['IsLargeLot'] = df['lot_sqft'] > 10000
print("\nFeature Engineering Preview:")
print(df[['year_built','Property Age','Price per SqFt','IsLargeLot']].head())
```
